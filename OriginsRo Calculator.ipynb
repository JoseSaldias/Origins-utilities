```python
status = 99
status_points_required = 0
print(
    '\n Pretrans point table \n\n',
    'Level', '\t\t  Status added', '   Requiered points \n', 
    '•••••••••••••••••••••••••••••••••••••••••••••••••'
    )
for i in range(status):
    print(
        '{} -> {}  '.format(i, i+1),
        '\t|      ', i//10 +2,
        '\t|      ', status_points_required, '\t  |'
    )
    status_points_required += i//10 +2
```

```python
base_str = 1
base_agi = 1
base_vit = 1
base_int = 1
base_dex = 1
base_luk = 1

def total_stats(level):
    total = 0
    for i in range(level):
        total += i//10 +2
    return total

def set_stat(status = 1, add = 0, level = 1, used = 0):
    required = 0
    for i in range(status + add -1):
        required += i//10 +2
    available = total_stats(level)
    remanent = available - required - used
    final = status + add
    print(
        '\nStatus points used: \t', required, 
        '\nRemaining points:\t', remanent)
    return final, required, remanent

level = 87
base_str, used, remaining = set_stat(status=base_str, add=48, level=level, used = 0)

print(
    '\nstr: ', base_str,
    '\nagi: ', base_agi,
    '\nvit: ', base_vit,
    '\nint: ', base_int,
    '\ndex: ', base_dex,
    '\nluk: ', base_luk,
)


```
